####1.2.1 Análisis de las columnas LEGAL_NAME, ORIGIN, SUPPLIER, EAN Y NUM

Se decide eliminar las siguientes columnas ya que no serán variables en nuestro estudio.

```python
df_productos = df_productos.drop(['LEGAL_NAME', 'ORIGIN', 'SUPPLIER','EAN','NUM','UNIT'], axis=1)
```

```python
print(df_productos['BRAND'].isnull().sum())
```

```python
df_productos['BRAND'].fillna('Desconocido', inplace=True)
```

####1.2.3 Análisis de la columna PRICE

```python
#Convertimos la columna 'PRICE' a float para ayudarnos a los análisis posteriores.
df_productos['PRICE'] = df_productos['PRICE'].str.replace(',', '.').astype(float)
```

```python
#Renombro la columna PRICE a 'PRECIO_UNIDAD'
df_productos = df_productos.rename(columns={'PRICE': 'PRECIO_UNIDAD'})
```

```python
print(df_productos['PRODUCT_CATEGORY'].dtype)
```

## 7.- Dataset de Eventos deportivos y feriados en ENERO 2021

https://blog.versus.es/calendario-de-los-principales-eventos-deportivos-de-enero-de-2021/
Se encontró información en la página de Apuestas Deportivas que se realiza de manera Online

Considerando que España es un país con muchos fanáticos del futbol, en Enero del año 2021 se realizó La "SUPERCOPA DE ESPAÑA".
Nos centraremos en las semifinales y final de la Supercopa de España 2020/2021 que se disputó el 13 de enero y 14 de enero las semifinales y la final el 17 de enero. Los dos finalistas de la Copa del Rey, Athletic y Real Sociedad, tendrán plaza en la Final Four contra Real Madrid y el Barcelona.


###7.1 Descripción del conjunto de datos

1.   FECHA_EVENTO: Fecha del evento
2.   EVENTO: Descripción del evento


### 7.2 Scrapping de la web

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Hacer la solicitud HTTP GET a la página web
url = 'https://blog.versus.es/calendario-de-los-principales-eventos-deportivos-de-enero-de-2021/'
response = requests.get(url)

# Crear un objeto BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Buscar todos los elementos <li> que contienen los eventos deportivos
event_elements = soup.find_all('li')

# Extraer las fechas y descripciones de los eventos correspondientes a enero
fechas = []
eventos = []
for evento in event_elements:
    # Obtener el texto del elemento
    texto_evento = evento.get_text(strip=True)

    # Verificar si el texto contiene una fecha en formato "X de Enero"
    if ' de Enero' in texto_evento:
        # Separar la fecha y la descripción
        fecha, descripcion = texto_evento.split(':', 1)

        # Agregar la fecha y la descripción a las listas correspondientes
        fechas.append(fecha.strip())
        eventos.append(descripcion.strip())

# Crear un DataFrame con los datos extraídos
data = {'FECHA': fechas, 'EVENTO': eventos}
df_eventos = pd.DataFrame(data)

# Imprimir el DataFrame completo
display(df_eventos)
```

##10.- Creación de la columna COD_EVENTO

Creamos la columna COD_EVENTO, donde le ponemos etiqueta a los eventos mas relevantes en el mes de enero del año 2021. Dónde tendrá las siguientes etiquetas:

*   'sin evento': 0,
*   'previo al día festivo': 1
*   'día festivo': 2

**Consideramos día fesitivo las siguientes fechas:**

2021-01-01 Año Nuevo

2021-01-06 Día de Reyes

2021-01-13 Primera semifinal Supercopa de España

2021-01-14 Segunda semifinal Supercopa de España

2021-01-17 Final Supercopa de España

Tambien consideramos cómo día festivo los días domingos, que no se abren la tienda.

Días domingos: 2021-01-3,2021-01-10,2021-01-17, 2021-01-24, 2021-01-31.

se asigna el valor de la etiqueta correspondiente a "día festivo" (2)

**Consideramos previo al día festivo:**
Día sabados: 2021-01-2, 2021-01-9, 2021-01-16, 2021-01-23, 2021-01-30 serán considerados como 'previo al día festivo'(1)

Si el valor de la columna "FECHA" es igual al 2021-01-5 de enero, al 2021-01-12 de enero, y 2021-01-16, se asigna el valor de la etiqueta correspondiente a "previo al día festivo" (1).

**Sin evento:**

En todos los demás casos, se asigna el valor de la etiqueta correspondiente a "sin evento" (0)





Si el valor de la columna "FECHA" es igual al 5 de enero o al 11 de enero, se asigna el valor de la etiqueta correspondiente a "previo al día festivo" (1). Si el valor de la columna "FECHA" es igual al 6 de enero, 12 de enero, 13 de enero o 17 de enero, se asigna el valor de la etiqueta correspondiente a "día festivo" (2). En todos los demás casos, se asigna el valor de la etiqueta correspondiente a "sin evento" (0).

```python
# Creamos la columna "COD_EVENTO" basada en la columna "FECHA"
import numpy as np

# Definimos las etiquetas de los eventos
etiquetas_evento = {
    'sin evento': 0,
    'previo al día festivo': 1,
    'día festivo': 2
}
# Definimos las condiciones y las etiquetas correspondientes
condiciones = [
    df_ventas_enero_merge['FECHA'].isin(['2021-01-01', '2021-01-06', '2021-01-13', '2021-01-14', '2021-01-17']),
    (df_ventas_enero_merge['FECHA'].dt.dayofweek == 6) | df_ventas_enero_merge['FECHA'].isin(['2021-01-02', '2021-01-09', '2021-01-16', '2021-01-23', '2021-01-30']),
    (df_ventas_enero_merge['FECHA'].dt.dayofweek == 5),
]

etiquetas = [etiquetas_evento['día festivo'], etiquetas_evento['previo al día festivo'], etiquetas_evento['sin evento']]

# Aplicamos las condiciones y etiquetas con numpy.select
df_ventas_enero_merge['COD_EVENTO'] = np.select(condiciones, etiquetas, default=etiquetas_evento['sin evento'])

# Asegurémonos de que la columna sea de tipo entero
df_ventas_enero_merge['COD_EVENTO'] = df_ventas_enero_merge['COD_EVENTO'].astype(int)


```
